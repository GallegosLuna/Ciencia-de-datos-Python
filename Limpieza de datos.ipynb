# Limpieza de datos con Pandas, algunos graficos

```python
import pandas as pd
import numpy as np
#Cargar datos
df=pd.read_csv("credits.csv") # Por defecto detecta las columnas por el encabezado del archivo
df2=pd.read_csv("titles.csv") # Por defecto detecta las columnas por el encabezado del archivo
df
```

```python
df2
```

## Intentaremos unir ambos data frames

```python
#Union de data frames de acuerdo con el indice (Columna que comparten ambas tablas)
leftdf=pd.merge(df,df2, on='id',how='left')
leftdf
```

#### Nota: Lo hice asi porque el data frame 1 tiene mas filas, por lo tanto asi perdemos menos datos originales, aunque vamos a ver muchos datos Nulos los cuales vamos a resulver

### Conocer el tipo de datos de todas las columnas

```python
df=leftdf
df.dtypes
```

### Conocer los datos nulos

```python
df.isnull().sum()
```

```python
df.describe().T
```

## Ahora que conocemos un poco mejor los datos, podemos descartar algunas columnas que no nos van a servir por tener muchos datos nulos y otras porque no importa su contenido.


```python
df=df.drop(['seasons', 'age_certification', 'character','imdb_id','imdb_votes'], axis=1)
df

```

```python
a=df.isnull().sum()
a
```

#### Es importante conocer la proporcion de los datos nulos para considerar eliminarlos o reemplazarlos. Con la finalidad de perder datos lo menos posible

```python
# Porcentaje de datos nulos
total_cells = np.product(df.shape)
total_missing = a.sum()

(total_missing/total_cells)*100
```

# Podemos eliminar ahora si los datos nulos porque solo representan un 0.5% del total

### Nota, aun asi no consideremos esto como una primera opcion. Para no perder todo nuestros datos, trataremos de reemplazarlos por el valor mas comun que aparece, o por la media para afectar lo menos posible

```python
df.describe().T
```

#### Podemos observar que IMDB_SCORE tiene media, al igual que los demas score. Para no afectar a nuestros datos reemplazaremos la moda en sus lugares correspondientes.


```python
values = {"imdb_score":6.466389 }
df.fillna(value=values, inplace=True)
values2 = {"tmdb_score":6.685284 }
df.fillna(value=values2, inplace=True)
values3 = {"tmdb_popularity":27.791778 }
df.fillna(value=values3, inplace=True)

```

```python
df.isnull().sum()
```

#### Como podemos ver, ahora no eliminaremos mucha informacion, por lo tanto podemos proceder a eliminar las filas con datos nulos de todo nuestro data frame

```python
df= df.dropna()
df.isnull().sum()
```

## Ahora filtraremos las columnas que van de acuerdo a nuestros intereses.


```python
df
```

#### Columnas que me sirven: title,	type,release_year, runtime,	genres,	production_countries,	imdb_score,	tmdb_popularity,tmdb_score

```python
df=df[["title", "type",'release_year','runtime','genres','imdb_score','tmdb_popularity','tmdb_score','name','genres']] 
df
# Me devuelve un dataframe solamente con esas columnas
```

# Algunas ocasiones tendremos que agrupar las cosas, por ejemplo aqui podemos agrupar tanto por actores como pelicula, genero.  En este caso se agrupara por peliculas.

```python
# AGRUPANDO Y AGREGANDO DATOS
peliculas=df.groupby("title").mean()# Agrupando por titulo y sacando la media de cada columna
peliculas
df=peliculas
df
```

### Aqui termina nuestra limpieza de datos, como vemos ya tenemos una base de datos mas entendible, mas trabajable.

### Aun asi podemos agregar columnas a este data frame en base a una transformacion de datos, por ejemplo el promedio de ambas calificaciones 

```python
# Agregando una columna de promedio, calculada en base a las calificaciones
def promedio(fila):
    return ((fila["imdb_score"]+fila["tmdb_score"])*.5)

df["Scorepromedio"] = df.apply(promedio, axis=1)
df
```

# Por ultimo exportaremos a un nuevo CSV este data frame ya trabajado, para posteriormente hacer graficas y visualizaciones.

```python
# ALMACENAMIENTO
df.to_csv("Peliculas.csv")
```
